```python
from collections import Counter
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from gensim.parsing.preprocessing import STOPWORDS
from gensim.utils import simple_preprocess
from gensim.corpora.dictionary import Dictionary
```

```python
def text_to_list(text):
    return [s for s in str(text).strip('\n').split('.')]
```

```python
def doc_to_text(src_dir):
    docs = []
    for fl in os.listdir(src_dir):
        if len(fl) == 29:
            full_text = textract.process(src_dir + fl)
            docs += text_to_list(full_text)
    return [chapter.replace('\\n', ' ') for chapter in docs]
```

```python
def text_to_words(text):
    return [[token for token in simple_preprocess(chapter) if token not in STOPWORDS] 
            for chapter in text]
```

```python
def corpus_to_list(corpus):
    oned_vec = []
    for doc in corpus:
        for d in doc:
            oned_vec.append(d)
    return oned_vec
```

```python
texts = doc_to_text(src_dir=src_dir)
```

```python
corpora = text_to_words(text=texts)
```

```python
dictionary = Dictionary(corpora)
```

```python
print(dictionary)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(doc) for doc in corpora]
```

```python
key_list = corpus_to_list(corpus=corpus)
```

```python
keys_sorted = sorted(key_list, key=lambda w: w[1], reverse=True)
```
