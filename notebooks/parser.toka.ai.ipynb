```python
from nltk import tokenize
from collections import Counter
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from gensim.parsing.preprocessing import STOPWORDS
from gensim.utils import simple_preprocess
from gensim.corpora.dictionary import Dictionary
```

```python
def text_to_list(text):
    return [s for s in str(text).strip('\n').split('.')]
```

```python
def doc_to_sentences(src_dir):
    docs = []
    for fl in os.listdir(src_dir):
        if len(fl) == 29:
            full_text = textract.process(src_dir + fl)
            docs += text_to_list(full_text)
    return [sentences.replace('\\n', ' ') for sentences in docs]
```

```python
my_stopwords = ['xa', 'xe', 'xc', 'iii', 'ii', 'iv', 'vi']
```

```python
def sentences_to_words(sentences, custom_stopwords=[]):
    wordnet_lemmatizer = WordNetLemmatizer()
    return [[wordnet_lemmatizer.lemmatize(token) 
             for token in simple_preprocess(sentence) 
             if token not in STOPWORDS and token not in custom_stopwords] 
            for sentence in sentences]
```

```python
def process_sentence(sent):
    return sent.lower
```

```python
def corpus_to_list(corpus):
    oned_vec = []
    for doc in corpus:
        for d in doc:
            oned_vec.append(d)
    return oned_vec
```

```python
sentences = doc_to_sentences(src_dir=src_dir)
```

```python
corpora = sentences_to_words(sentences, my_stopwords)
```

```python
dictionary = Dictionary(corpora)
```

```python
print(dictionary)
```

```python
print(dictionary.token2id)
```

```python
corpus = [dictionary.doc2bow(doc) for doc in corpora]
```

```python
key_list = corpus_to_list(corpus=corpus)
```

```python
keys_sorted = sorted(key_list, key=lambda w: w[1], reverse=True)
```

```python
words = corpus_to_list(corpora)
```

```python
words_sorted = sorted(words, key=lambda w: w[1], reverse=True)
```

```python
Counter(words_sorted).most_common(len(words_sorted))
```

```python
word_counts = {}
for word_id, word_count in keys_sorted:
    word = dictionary.get(word_id)
    if word in word_counts.keys():
        word_counts[word] += word_count
    else:
        word_counts[word] = word_count
    
print(word_counts)    
```
